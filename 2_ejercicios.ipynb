```python
# ejercicio 11
import random

def generar_matriz_aleatoria(N):
    matriz = []
    for i in range(N):
        fila = []
        for j in range(N):
            fila.append(random.randint(1, 9))
        matriz.append(fila)
    return matriz

def imprimir_matriz(matriz):
    for fila in matriz:
        fila_formateada = ' '.join(f'{elemento:2}' for elemento in fila)
        print(fila_formateada)

# Generar matriz 3x3
print("Matriz de 3x3:")
m_3 = generar_matriz_aleatoria(3)
imprimir_matriz(m_3)
```
